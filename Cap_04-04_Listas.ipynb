#Funções Built-In

```python
# Criando uma lista
lista_numeros = [10, 20, 50, -3.4]
```

```python
# Função len() retorna o comprimento da lista
len(lista_numeros)
```

```python
# Função max() retorna o valor máximo da lista
max(lista_numeros)
```

```python
# Função min() retorna o menor valor
min(lista_numeros)
```
