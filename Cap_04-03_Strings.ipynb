##Indexando Strings

```python
# Atribuindo uma string
a = 'Data Science Academy'
```

```python
print(a)
```

Indexação em python começa em 0!!!

```python
# Primeiro elemento da string
a[0]
```

```python
a[1]
```

```python
a[2]
```

```python
a[3]
```

Pode-se usar : para executar um slicing que faz a leitura de tudo até um ponto designado. Por exemplo:

```python
# Retorna todos os elementos da string, começando pela posição 0 até o fim da string
a[0:]
```

```python
a[1:]
```

```python
# A string original permanece inalterada
a
```

```python
# Retorna tudo até a posição 3 (mas não inclusa)
a[:3]
```

```python
# Indexação negativa: lê de trás para frente
a[-1]
```

```python
# Retorna tudo, exceto o último caractere
a[:-1]
```

Pode-se usar :: para fatiar a string (o padrão é 1). Se preferirmos outro intervalo, é só indicá-lo após o ::

```python
a[::]  # Imprime todas as letras
```

```python
a[::1]  # Imprime a string pulando 1 caractere, ou seja, imprime tudo
```

```python
a[::2]  # Imprime a cada 2 caracteres, ou seja, um sim - um não
```

```python
a[::-1]
```
