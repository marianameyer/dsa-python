#**Capítulo 5 - Métodos**

Tudo em python é objeto. E cada objeto tem métodos e atributos.



*   Atributos são propriedade, características dos objetos.
*   Métodos são funções com ações que podem ser executadas nos objetos.



```python
# Criando uma lista
lista = [100, -2, 12, 65, 0]
```

```python
type(lista)
```

```python
# Verificando métodos e atributos
lista
```

```python
# Usando um método do objeto lista
lista.append(100)
```

```python
# Imprimindo a lista
print(lista)
```

```python
# Usando um método do objeto lista
lista.count(100)
```

```python
# A função help() explica como utilizar cada método de um objeto
help(lista.count)
```

```python
# A função dir() mostra todos os métodos e atributos de um objeto
dir(lista)
```

```python
frase = 'Isso é uma string'
```

```python
type(frase)
```

```python
# Verificando métodos e atributos
frase
```

```python
# O método de um objeto pode ser chamado de dentro de uma função, como print()
print(frase.split())
```
