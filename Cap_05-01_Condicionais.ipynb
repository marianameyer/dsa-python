##Condicional IF:

```python
# Condicional if
if 5 > 2:
  print('A sentença é verdadeira.')
```

```python
# Condicional if...else
if 5 < 2:
  print('A sentença é verdadeira.')
else:
  print('A sentença é falsa!')
```

```python
# Condicional if...else com variável
dia = 'Terça'
if dia == 'Segunda':
  print('Hoje fará sol!')
else:
  print('Hoje vai chover!')
```

```python
# Podemos usar o operador elif para validar mais de uma condição
if dia == 'Segunda':
  print('Hoje fará sol!')
elif dia == 'Terça':
  print('Hoje vai chover!')
else:
  print('Sem previsão de tempo para o dia selecionado!')
```

- AND: retorna True se ambas forem verdadeiras;
- OR: retorna True se uma das declarações for verdadeira;
- NOT: Inverte o resultado.

```python
# Operador AND
numero = 4
if (numero > 2) and (numero % 2 == 0):
  print('Isso está sendo impresso pois ambas as afirmações são verdadeiras.')
```

```python
# Operador AND
numero = 4
if (numero > 5) and (numero % 2 == 0):
  print('Isso está sendo impresso pois ambas as afirmações são verdadeiras.')
else:
  print('Isso está sendo impresso pois uma das afirmações é falsa.')
```

```python
# Operador OR
numero = 4
if (numero > 5) or (numero % 2 == 0):
  print('Isso está sendo impresso pois uma das afirmações é verdadeira.')
```

```python
# Operador NOT
numero = 4
if not(numero > 5) or (numero % 2 == 0):
  print('Isso está sendo impresso pois ambas as afirmações são verdadeiras.')
else:
  print('Isso está sendo impresso pois uma das condições é falsa.')
```

```python
# Operador AND, OR e NOT
numero = 4
if (not(numero > 5) and (numero % 2 == 0)) or (numero == 4):
  print('Isso está sendo impresso pois as duas primeiras condições são verdadeiras ou a terceira é verdadeira.')
```

```python
# Exemplo com uso de variáveis
disciplina = 'Data Science'
nota_final = 70
if disciplina == 'Data Science' and nota_final >= 70:
  print('Você foi aprovado!')
else:
  print('Lamento, acho que você precisa estudar mais!')
```

```python
# Usando mais de uma condição na cláusula if
disciplina = 'Data Science'
nota_final = 60
if disciplina == 'Data Science' and nota_final >= 70:
  print('Você foi aprovado!')
else:
  print('Lamento, acho que você precisa estudar mais!')
```

```python
# Usando mais de uma condição na cláusula if e introduzindo placeholders
disciplina = 'Data Science'
nota_final = 90
semestre = 2
if disciplina == 'Data Science' and nota_final >= 80 and semestre != 1:
  print('Você foi aprovado em %s com média final %r!' %(disciplina, nota_final))
else:
  print('Lamento, acho que você precisa estudar mais!')
```
