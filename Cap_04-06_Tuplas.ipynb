#*Capítulo 4 - Tuplas*

##Trabalhando com Tuplas

```python
# Criando uma tupla
tupla1 = ('Geografia', 23, 'Elefantes', 9.8, 'Python')
```

```python
# Imprimindo a tupla
tupla1
```

```python
# Tuplas não suportam append()
tupla1.append('Chocolate')
```

```python
# Tuplas não suportam deletar um item específico
del tupla1['Elefantes']
```

```python
# Tuplas podem ter apenas um item
tupla1 = ('Chocolate')
```

```python
tupla1
```

```python
tupla1 = ('Geografia', 23, 'Elefantes', 9.8, 'Python')
```

```python
tupla1[0]
```

```python
# Verificando o comprimento da tupla
len(tupla1)
```

```python
# Slicinc, da mesma forma como em listas
tupla1[1:]
```

```python
tupla1.index('Elefantes')
```

```python
# Tuplas não suportam atribuição de item
tupla1[1] = 21
```

```python
# Deletando a tupla
del tupla1
```

```python
tupla1
```

```python
# Criando uma tupla
t2 = ('A', 'B', 'C')
```

```python
t2
```

```python
# Tuplas não suportam atribuição de item
t2[0] = 'D'
```

```python
# Usando a função list() para converter um tupla em lista
lista_t2 = list(t2)
```

```python
lista_t2
```

```python
lista_t2.append('D')
```

```python
lista_t2
```

```python
# Usando a função tuple para converter a lista em tupla
t2 = tuple(lista_t2)
```

```python
t2
```
